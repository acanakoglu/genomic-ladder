# Convert input data to desired format 

```python
import pickle
import pandas as pd
import numpy as np
from collections import Counter
```

```python
regiondata_url = "./gmql_data/Datasets/gs_breast_finale.pickle"
regiondata = pickle.load(open(regiondata_url, "rb"))
```

```python
regiondata
```

Get labels from the index

```python
lbls = regiondata.index.get_level_values('manually_curated__tissue_status')
len(lbls)
```

```python
lbls= ["1" if x=="tumoral" else "0" for x in lbls]
for i in range(len(lbls)):
    lbls[i] = int(lbls[i])
```

Dump the unnecessery columns

```python
regiondata = regiondata.normalized_count
regiondata.index = regiondata.index.get_level_values(-1)
regiondata.head()
```

```python
regiondata = regiondata.fillna(regiondata.mean())
regiondata = regiondata.dropna(axis=1, how='all')
```

```python
regiondata
```

```python
regiondata = (regiondata - regiondata.mean())
regiondata = regiondata / (regiondata.max() - regiondata.min())
regiondata
```

Save base data to pickel file

```python
pickle_file = "./gmql_data/Datasets/breast_data_all_w_labels.p"

try:
    
    f = open(pickle_file, 'wb')
    save = {
    'data': regiondata,
    'labels': lbls,

    }
    pickle.dump(save, f, pickle.HIGHEST_PROTOCOL)
    f.close()
except Exception as e:
    print('Unable to save data to', pickle_file, ':', e)
    raise

    
    
    
```

# Split the data for train and test

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import StratifiedShuffleSplit
import pickle
import numpy as np
```

```python
pickle_file = "./gmql_data/Datasets/breast_data_all_w_labels.p"
with open(pickle_file, 'rb') as f:
    save = pickle.load(f)
    data = save['data']
    labels = save['labels']
    del save  # hint to help gc free up memory
```

Set labels to y parameter

```python
x = np.array(data)
y = np.array(labels)
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
sss.get_n_splits(x, y)
sss
```

```python
for train_index, test_index in sss.split(x, y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

Create validation dataset from training data

```python
x = x_train
z = y_train

sss = StratifiedShuffleSplit(n_splits=3, test_size=0.25, random_state=0)
sss.get_n_splits(x, z)
sss
```

```python
for train_index, valid_index in sss.split(x, z):
    x_train, x_valid = x[train_index], x[valid_index]
    z_train, z_valid = z[train_index], z[valid_index]
```

Save the data for future use

```python
pickle_file = "./gmql_data/Datasets/breast_data_all_divided_4.pickle"

try:
  f = open(pickle_file, 'wb')
  save = {
    'train_dataset': x_train,
    'train_labels': z_train,
    'test_dataset': x_test,
    'test_labels': y_test,
    'valid_dataset': x_valid,
    'valid_labels': z_valid,
    }
  pickle.dump(save, f, pickle.HIGHEST_PROTOCOL)
  f.close()
except Exception as e:
  print('Unable to save data to', pickle_file, ':', e)
  raise
```
